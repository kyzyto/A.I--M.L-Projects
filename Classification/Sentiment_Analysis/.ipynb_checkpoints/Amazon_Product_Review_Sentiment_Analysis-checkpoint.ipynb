### ML: Split the dataset into Train and Test

[Reference](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html)

```python
from sklearn.model_selection import StratifiedShuffleSplit

print("Before Drop {}".format(len(data)))

print("Number of null rows in reviews.rating {}".format(data["reviews.rating"].isnull().sum()))

print("After Drop {}".format(len(data.dropna(subset=["reviews.rating"]))))
```

```python
# Remove all the NAN in the column reviews.rating
data_after = data.dropna(subset=["reviews.rating"])
print("Type before conversion %s" % str(data_after["reviews.rating"].dtypes))

data_after["reviews.rating"] = data_after["reviews.rating"].astype(int) #converts from float to int
print("Type after conversion %s" % str(data_after["reviews.rating"].dtypes))
```

[Reference](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html)

Since goal is to train a sentiment analysis classifier and the majority of reviews are positive (i.e 5 stars), we need to use Stratified ShuffleSplit on the reviews score to ensure that we don't train the classifier on imbalanced data

```python
split = StratifiedShuffleSplit(n_splits = 5, test_size=0.2)

for train_index, test_index in split.split(data_after, data_after["reviews.rating"]):
    strat_train = data_after.reindex(train_index)
    strat_test = data_after.reindex(test_index)

print(len(strat_train))
print(len(strat_test))
print(strat_test["reviews.rating"].value_counts())
print(strat_test["reviews.rating"].value_counts()/len(strat_test))
```

### Sentiment Analysis

```python
def sentiments(rating):
    if (rating == 5) or (rating == 4):
        return ("Positive ðŸ˜Š")
    elif rating == 3:
        return "Neutral"
    else:
        return ("Negative ðŸ˜ ")
        
strat_train["Sentiment"] = strat_train["reviews.rating"].apply(sentiments)
print(strat_train["Sentiment"][:10])

strat_test["Sentiment"] = strat_train["reviews.rating"].apply(sentiments)
print(strat_test["Sentiment"][:10])
```
